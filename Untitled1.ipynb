```python
# import the necessary packages
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Read the image and convert to grayscale
image=cv2.imread("lanes.jpg")
img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.axis('off')
plt.title('gray_image')
plt.imshow(img,'gray')
```

```python
# Find the edges using canny edge detector
gaussian_blur=cv2.GaussianBlur(img,(3,3),0)
edges1 = cv2.Canny(gaussian_blur, 120, 150)
plt.imshow(edges1,'gray')
```

```python
# Detect the points that form the line
lines=cv2.HoughLinesP(edges1,1,np.pi/180,threshold=80,minLineLength=50,maxLineGap=250)
```

```python
# Draw lines on the image
for line in lines:
    x1,x2,y1,y2=line[0]
    cv2.line(img,(x1,y1),(x2,y2),(255,0,0),3)
# Display the results
plt.imshow(img,'gray') 
plt.axis('off')
```
